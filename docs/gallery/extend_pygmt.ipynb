```python
from antarctic_plots import maps, fetch, utils

imagery = fetch.imagery()

region = [150e3, 550e3, -1350e3, -1100e3]

fig = maps.plot_grd(
    imagery,
    plot_region=region,
    image=True,
    inset=True,
    title="LIMA imagery",
    grid_lines=True,
    x_annots=10,  # plot 10 degree longitude lines
    y_annots=2,  # plot 2 degree latitude lines
)

# use standard PyGMT methods
# add surface elevation contours
surface = fetch.bedmap2(layer="surface")
fig.grdcontour(grid=surface, interval=500, pen="thin,black")

fig.show()
```

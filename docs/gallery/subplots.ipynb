# Create subplots

```python
from antarctic_plots import maps, fetch

region = [-1800e3, -1000e3, -1200e3, -700e3]

bed = fetch.bedmap2(layer="bed")
gravity = fetch.gravity(type="FA", region=region)

fig = maps.plot_grd(
    bed,
    plot_region=region,
    grd2cpt=True,
    cmap="batlowW",
    coast=True,
    cbar_label="Bedmap2 bed elevation (m)",
    grid_lines=True,
)

fig = maps.plot_grd(
    gravity,
    plot_region=region,
    grd2cpt=True,
    cmap="viridis",
    coast=True,
    cbar_label="Free-air gravity (mGal)",
    origin_shift="yshift",
    fig=fig,
    inset=True,
    inset_pos="TR",
    grid_lines=True,
)

fig.show()
```
